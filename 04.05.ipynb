```python
for n in range(1000, 0, -1):
    b = bin(n)[2:]
    ost = b.count('1') % 3
    if ost == 0:
        r = b + b[:2]
    else:
        r = bin(3 * ost)[2:] + b
    if int(r, 2) <= 60:
        print(int(r, 2), n)
        break
```

```python
s = list(map(int, open('17.txt').readlines()))

m3 = max([i for i in s if len(str(abs(i))) == 3])

c = 0
m = 0
for i in range(len(s) - 1):
    if (len(str(abs(s[i]))) == 3 or len(str(abs(s[i + 1]))) == 3) and \
        (len(str(abs(s[i]))) != len(str(abs(s[i+1])))) and \
        (s[i] + s[i+1]) <= m3:
        c += 1
        m = max(0, s[i] + s[i+1])

print(c, m)
```
