```python
def to3(n):
    r = ''
    while n > 0:
        r += str(n % 3)
        n //= 3
    r = r[::-1]
    return r

def f(n):
    r = to3(n)
    if n % 2 == 0:
        r += r[-2:]
    else:
        r += to3(sum([int(i) for i in r]))
    return int(r, 3)


print(
    f(10), f(11)
)
ans = (10, f(10))
for n in range(10, 1000):
    r = f(n)
    if ans[1] > r:
        ans = (n, r)
print(ans)
```

```python
alph = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'.lower()
L = alph.find('l')
V = alph.find('v')

f = 0
for q in range(V + 1, 37):
    p2 = int('danov', q)
    for p in range(L + 1, 37):
        for x in alph[:p]:
            p1 = int('ale'+x, p)
            if (p1 + p2) % 2023 == 0:
                print((p1+p2)//2023)
                f = 1
                break
        if f:
            break
    if f:
        break
```
